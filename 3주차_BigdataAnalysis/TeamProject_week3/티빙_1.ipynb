## 4. 총평

- 성별에 대한 관심도의 증감 여부는 작품들. 즉, 콘텐츠들의 인기도에 따라 유동적으로 변하는 것으로 보임  
    -> 양질의 콘텐츠를 제작하는 것에 집중  
  
  
< STP >


- Segmentaion (소비자층 구분)
    - 성별에 따른 장르의 관심도 등락폭은 서비스하는 작품에 따른 변동성이 큰 것으로 볼 수 있음 (비교해 봤을 때도)
    - 성별보다는 연령대에 맞춘 마케팅 전략을 통한 성과를 노려봅시다.  
    

- Targeting (핵심 소비자층)
    - 관심도가 하락하는 젊은 세대
    - 'A세대' : 높은 구매력으로 여유로운 장밋빛 인생 2막을 시작한 50~69세 시니어(Senior)를 의미  
        -> 유행에 민감하지 않고, 높은 충성도/구매력, 새 IT 기기 구매 등  
        -> ex) 임영웅의 음악 플랫폼 상위 랭크, 팬들 조공 문화, 단독 콘서트 매진  


- Positioning (국내시장, A세대)
    - '국내, A세대를 위한 콘텐츠 및 서비스의 강자' 라는 콘셉으로 방향 설정
    - CJ 계열, 2010년부터 누적된 국내 고객 정보 확보, JTBC, 네이버의 합류 => 국내 시장에서의 강점
    - 강점을 살려 ->  ex) 실제, 티빙에서 임영웅의 서울 공연을 중계 해주기도 했다고 함
