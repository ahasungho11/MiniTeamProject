## 1. 장르마다의 성별 데이터

```python
# 모듈 로딩
import pandas as pd
import warnings
warnings.simplefilter("ignore")
import matplotlib.pyplot as plt

# 폰트
from matplotlib import font_manager as fm, rc

font_path = './Fonts/malgun.ttf'
font_name = fm.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

# 파일 경로
DIR = '../Data_1/'
FILE_ = [ DIR+'tving_movie_M.xlsx', DIR+'tving_movie_F.xlsx', DIR+'tving_drama_M.xlsx',
         DIR+'tving_drama_F.xlsx', DIR+'tving_animation_M.xlsx', DIR+'tving_animation_F.xlsx',
         DIR+'tving_entertainment_M.xlsx', DIR+'tving_entertainment_F.xlsx' ]
# df1 -> df8 : 영화(남/여), 드라마(남/여), 애니메이션(남/여), 예능(남/여) 순서
```

```python
# 파일 로딩 -> DF 저장
df =[]
for i in FILE_:
    df.append( pd.read_excel(i, skiprows=6) )
```

### 1) 장르별 남자들의 자료

```python
# 남자의 자료들만 모아서 DF로 만들기
df[0].insert(2, '드라마_남자', df[2]['tving'])
```

```python
df[0] = pd.concat([df[0],df[4]['tving']],axis=1)
df[0] = pd.concat([df[0],df[6]['tving']],axis=1)
```

```python
df[0].columns
```

```python
# 컬럼명 맞추기
df[0].columns = ['날짜', '영화', '드라마', '애니메이션', '예능']
```

```python
male_type = df[0]
male_type
```

### 2) 장르별 여자들의 자료

```python
# 여자의 자료들만 모아서 DF로 만들기
df[1].insert(2, '드라마_여자', df[3]['tving'])
```

```python
df[1] = pd.concat([df[1],df[5]['tving']],axis=1)
df[1] = pd.concat([df[1],df[7]['tving']],axis=1)
```

```python
df[1].columns
```
