```python
import pandas as pd

DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH+'sports_Seoul_2020.xlsx'
FILE_NAME
```

```python
# 데이터프레임 저장 - 서울 체육시설 현황(2020)
sports_Seoul = pd.read_excel(FILE_NAME, sheet_name=None, skiprows=1)
# 각 시트의 이름이 key, 시트의 내용이 value인 딕셔너리
# skiprows=1로 두었다 => 기존의 컬럼명을 날리고 아래 행을 위로 올려서 컬럼명으로 쓰기 위해서 

sports_Seoul                                              
```

```python
# 변수 정리
# sports_Seoul : 파일에 포함된 모든 시트를 불러온 것
# Seoul_list_all : 모든 시트의 이름을 뽑아서 리스트에 담은 것
# Seoul_list : 정보를 뽑을 시트 이름만을 모아 놓은 리스트
# Seoul_condition : 엑셀 파일 내에 있는 요약된 정보 시트명
# Seoul_col : 해당 시트에서 뽑을 컬럼명만을 모아 놓은 리스트
```

```python
Seoul_condition = sports_Seoul[Seoul_list_all[2]]
# 전반적인 현황판은 따로 뽑아두고 (어딘가에 활용)
# 차후에 행 전체 값이 NaN일 경우, 날리든지 해야함
Seoul_condition
```

```python
# 각 시트별로 필요한 컬럼명을 리스트로 뽑아내기

sports_Seoul[Seoul_list[0]].columns
# 젤 위에 날리고 컬럼명 새로 적어야 겠는데? => 불러오는 파라미터에서 skiprows=1로 해결

Seoul_col = sports_Seoul[Seoul_list[0]].columns[2:12]
# Seoul_col = ['시설명'] #,'주소'] #,'시설명','관리주체','부지면적'] #,'준공연도']
# 각 시트마다 뺴낼 컬럼들의 이름을 뽑아서 리스트로 담아
# 컬럼 리스트를 뽑기위해 추출할 가장 앞 시트인 축구장의 시트에 해당하는 컬럼명을 뽑은 것

# Seoul_col
# # key - 시트명
# # values - 데이터 프레임
# # 한수씨 확인했던 것 숙지 해보기

target_col = ['시군구','시설명','소유기관','부지면적']

# 확인
# cols = []
# for sheetName in Seoul_list:
#     df = sports_Seoul[sheetName]
#     cols.append(set(target_col).issubset(set(df.columns)))
    
# pd.DataFrame({
#     'sheet name' : Seoul_list,
#     'cols' : cols
# })
```

```python
# 변수 정리
# sports_Seoul : 파일에 포함된 모든 시트를 불러온 것
# Seoul_list_all : 모든 시트의 이름을 뽑아서 리스트에 담은 것
# Seoul_list : 정보를 뽑을 '시트 이름'만을 모아 놓은 리스트
# Seoul_condition : 엑셀 파일 내에 있는 요약된 정보 시트명
# Seoul_col : 해당 시트에서 뽑을 '컬럼 이름'만을 모아 놓은 리스트
#             -> 한 시트 기준으로 슬라이싱으로 뽑았다가, 일부 컬럼명만 직접 입력으로 바꿈
```

```python
target_col = ['시군구','시설명','소유기관','부지면적']

Seoul_data = sports_Seoul[Seoul_list[0]]              # 시트를 읽음
Seoul_data = Seoul_data[target_col]  

df3 = pd.DataFrame(columns = Seoul_data.columns)

for spot in Seoul_list:                          # spot : 정보를 뽑을 시트를 하나씩 뽑아서 spot에 담음
    Seoul_data = sports_Seoul[spot]              # 시트를 읽음
    Seoul_data = Seoul_data[target_col]          # 읽은 시트 DF에서 필요한 컬럼넣어 데이터를 뽑아
    
    df3 = pd.concat([df3, Seoul_data], )           # pd2제목 -> pd2제목+Seoul_Data -> pd2제목+Seoul_Data+Seoul_Data
```

```python
df4 = df3.dropna(how='all')  # 모든 행이 NaN인 부분은 날리고 df4로 선언
df4
```

```python
mask=[ isinstance(data, int) for data in df4['시설명'] ]
mask
```
