```python
# 연령별
# 1) 1살-8살 (학창시절 이전)  &  2019-21 (3개년)
query11 = """
select a.연도, sum(a.`합계(계)`) as `1-8살(명)`
from 전국_유형_연령 as a
where (나이 >= 1 and 나이 <= 8)
group by 연도;
"""
```

```python
# 2) 9살-24살 (학창시절)  &  2019-21 (3개년) 
query12 = """
select sum(a.`합계(계)`) as `9-19살(명)`
from 전국_유형_연령 as a
where (나이 >= 9 and 나이 <= 24)
group by 연도;
"""

# 3) 25살(20대 중반) - 49살  &  2019-21 (3개년) 
query13 = """
select sum(a.`합계(계)`) as `25-49살(명)`
from 전국_유형_연령 as a
where (나이 >= 25 and 나이 <= 49)
group by 연도;
"""

# 4) 50대 이후  &  2019-21 (3개년)
query14 = """
select sum(a.`합계(계)`) as `50살 이후(명)`
from 전국_유형_연령 as a
where (나이 >= 50 )
group by 연도;
"""
```

```python
# 조회하여 줄단위로 읽기
def mysql1(qqq):
    conn = pymysql.connect(host='localhost', user='root', password='fkfk2rj2@!##H',
                            db='teamproject', charset='utf8')
    curs = conn.cursor()

    curs.execute(qqq)
    rows = curs.fetchall()
#     for row in rows:
#         print(row)
    conn.commit()
    conn.close()
    
    return list(rows)
```

```python
def sevenYear_graph():
    sevenYear = mysql1(query1)

    year = []
    dis = []
    pop = []
    for i in range(len(sevenYear)):
        year.append(sevenYear[i][0])
        dis.append(sevenYear[i][1])
        pop.append(sevenYear[i][2])

    x = year
    y1 = dis
    y2 = pop

    # 장애인수 증가 그래프
    fig, axes = plt.subplots(1,2, figsize=(20,10))
    plt.subplot(1,2,1)
    # plt.figure(figsize=(18,6))
    plt.plot(x,y1, 'blue', marker='s', label='장애인증가수(명)')
    plt.xticks(x, fontsize =15)
    plt.yticks(fontsize =15)

    plt.ylabel('증가수', fontsize='20')
    plt.xlabel('연도', fontsize='20')
    plt.legend(loc = 2, fontsize=15)
    plt.title('< 장애인구 변화 추이 7개년(2015-2021년) >', fontsize='20')
    plt.style.use("ggplot")

    # 전국인구수 증가 그래프
    plt.subplot(1,2,2)
    plt.plot(x,y2, 'red', marker='s', label='전국인구증가수(명)')
    plt.xticks(x, fontsize =15)
    plt.yticks(fontsize =15)

    plt.ylabel('증가수', fontsize='20')
    plt.xlabel('연도', fontsize='20')
    plt.legend(loc = 2, fontsize=15)
    plt.title('< 전국인구 변화 추이 7개년(2015-2021년) >', fontsize='20')
    plt.style.use("ggplot")

    plt.show()
```

```python
def lowHigh_df():    
    lowHigh = mysql1(query2)

    assort = []
    dis = []
    weak = []
    
    for i in range(len(lowHigh)):
        assort.append(lowHigh[i][0])
        dis.append(lowHigh[i][1])
        weak.append(lowHigh[i][2])

    x = assort
    y1 = dis
    y2 = weak

    data1 = pd.Series(y1)
    data2 = pd.Series(y2)

    lowHigh_df = pd.concat([data1, data2], axis=1).T
    lowHigh_df.columns = x

    inxName=['심한 장애', '심하지 않은 장애']

    lowHigh_df.index = inxName
    lowHigh_df = lowHigh_df.T
    return lowHigh_df    
```

```python
def age_graph_data():

    year = []
    age1 = []
    age2 = []
    age3 = []
    age4 = []

    # 1살-8살
    age_graph = mysql1(query11)
    age1.append(int(age_graph[0][1]))
    age2.append(int(age_graph[1][1]))
    age3.append(int(age_graph[2][1]))

    # 9살-24살
    age_graph = mysql1(query12)
    age1.append(int(age_graph[0][0]))
    age2.append(int(age_graph[1][0]))
    age3.append(int(age_graph[2][0]))

    # 25살-49살
    age_graph = mysql1(query13)
    age1.append(int(age_graph[0][0]))
    age2.append(int(age_graph[1][0]))
    age3.append(int(age_graph[2][0]))

    # 50살 이상
    age_graph = mysql1(query14)
    age1.append(int(age_graph[0][0]))
    age2.append(int(age_graph[1][0]))
    age3.append(int(age_graph[2][0]))
```

```python
def age_graph_draw():
    # 색깔 리스트 만들기
    colors = sns.color_palette("muted", 4)
    
    # 원그래프 그리기
    fig, axes = plt.subplots(1,3, figsize=(20,10))
    plt.subplot(1,3,1)
    labels = ['1-8살', '9-24살', '25-49살', '50살 이상']
    # plt.figure(figsize=(10,8))
    plt.pie(age1, labels=labels, colors = colors, autopct='%.1f%%', textprops={'fontsize':16}, explode=[0, 0, 0, 0.1])
    plt.title('총 장애인수 대비 연령대별 비율(2019)')
    plt.legend(loc = 2, fontsize=13, shadow=True, facecolor='white')

    plt.subplot(1,3,2)
    labels = ['1-8살', '9-24살', '25-49살', '50살 이상']
    # plt.figure(figsize=(10,8))
    plt.pie(age2, labels=labels, colors = colors, autopct='%.1f%%', textprops={'fontsize':16}, explode=[0, 0, 0, 0.1])
    plt.title('총 장애인수 대비 연령대별 비율(2020)')
#     plt.legend(loc = 2, fontsize=13, shadow=True, facecolor='white')

    plt.subplot(1,3,3)
    labels = ['1-8살', '9-24살', '25-49살', '50살 이상']
    # plt.figure(figsize=(10,8))
    plt.pie(age3, labels=labels, colors = colors, autopct='%.1f%%', textprops={'fontsize':16}, explode=[0, 0, 0, 0.1])
    plt.title('총 장애인수 대비 연령대별 비율(2021)')
#     plt.legend(loc = 2, fontsize=13, shadow=True, facecolor='white')

    plt.show()
```
